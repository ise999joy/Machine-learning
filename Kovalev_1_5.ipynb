```python
# Лабораторна робота 1
# Вступ до машинного навчання. Елементи лінійної алгебри.

# Завдання 1.
# Створіть одновимірний масив (вектор) зі зворотнім порядком перших 10 натуральних чисел та виведіть його значення.

import numpy as np

arr = np.arange(1, 11)

arr_reversed = arr[::-1]

print("Одновимірний масив:", arr_reversed)
```

```python
# Завдання 2.
# Створіть двовимірний масив (матрицю) розміром 3x3,
# заповніть його послідовністю чисел, де кожне наступне число менше попереднього на 3,
# та виведіть його значення.

import numpy as np

start = 100

seq = np.arange(start, start - 9*3, -3)

matrix = seq.reshape(3, 3)

print("Матриця 3x3:")
print(matrix)
```

```python
# Завдання 4.
# Створіть масив розміром 3x3, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та обчисліть добуток всіх елементів.

import numpy as np

matrix = np.random.rand(3, 3)
product = np.prod(matrix)

print("Матриця 3x3:")
print(matrix)
print("Добуток усіх елементів:", product)
```

```python
# Завдання 5.
# Створіть два одновимірних масиви розміром 6, заповніть їх випадковими цілими числами в діапазоні від -20 до 20
# та виконайте на них поелементні операції додавання, віднімання та множення.

import numpy as np

a = np.random.randint(-20, 21, 6)
b = np.random.randint(-20, 21, 6)

print("Масив A:", a)
print("Масив B:", b)

print("A + B:", a + b)
print("A - B:", a - b)
print("A * B:", a * b)
```

```python
# Завдання 6.
# Створіть дві матриці 2x2, заповніть їх довільними цілими числами від 6 до 10 та знайдіть їх добуток.

import numpy as np

A = np.random.randint(6, 11, (2, 2))
B = np.random.randint(6, 11, (2, 2))
C = np.dot(A, B)

print("Матриця A:")
print(A)
print("Матриця B:")
print(B)
print("Добуток A * B:")
print(C)
```

```python
# Завдання 7.
# Створіть дві матриці розміром 3x2 та 2x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 8 та перемножте їх між собою.

import numpy as np

A = np.random.randint(1, 9, (3, 2))
B = np.random.randint(1, 9, (2, 2))
C = np.dot(A, B)

print("Матриця A (3x2):")
print(A)
print("Матриця B (2x2):")
print(B)
print("Добуток A * B (3x2 * 2x2 = 3x2):")
print(C)
```

```python
# Завдання 8.
# Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть її обернену матрицю.

import numpy as np

A = np.random.randint(1, 11, (2, 2))
inv_A = np.linalg.inv(A)

print("Матриця A:")
print(A)
print("Обернена матриця A⁻¹:")
print(inv_A)
```

```python
# Завдання 9.
# Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 0 до 1 та транспонуйте її.

import numpy as np

A = np.random.randint(0, 2, (3, 3))
AT = A.T

print("Матриця A:")
print(A)
print("Транспонована матриця Aᵀ:")
print(AT)
```

```python
# Завдання 10.
# Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть матрицю випадковими дійсними числами від 0 до 1,
# а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть добуток матриці на вектор.

import numpy as np

A = np.random.rand(3, 4)
v = np.random.randint(1, 11, (4, 1))
result = np.dot(A, v)

print("Матриця A (3x4):")
print(A)
print("Вектор v (4x1):")
print(v)
print("Добуток A * v (3x1):")
print(result)
```

```python
# Завдання 15.
# Створіть матрицю розміром 7x7, заповніть її випадковими дійсними числами в діапазоні від 0 до 1
# та знайдіть вектор-стовпчик, що містить найбільший елемент у кожному рядку матриці.

import numpy as np

A = np.random.rand(7, 7)
max_in_rows = np.max(A, axis=1).reshape(7, 1)

print("Матриця A:")
print(A)
print("Вектор-стовпчик найбільших елементів у рядках:")
print(max_in_rows)
```
